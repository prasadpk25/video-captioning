```python
from datasets import load_dataset

# Load the full dataset from Hugging Face Hub
ds = load_dataset("MBZUAI/VideoInstruct-100K", split="train")

print(ds)
print(ds[0])
```

```python
import pandas as pd

# Convert to pandas DataFrame
df = ds.to_pandas()

# Save as JSON or CSV
df.to_json("VideoInstruct_100K.json", orient="records", lines=True)
df.to_csv("VideoInstruct_100K.csv", index=False)

print("✅ Saved dataset locally!")
```
